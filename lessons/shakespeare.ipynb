# Quantifying Shakespeare
1. Read all of shakespeare's works into a list of strings.
2. Count the frequencies of the letters used.
3. Visualize the results of the frequencies

The dataset we will use is: <https://bit.ly/shakespeare-txt>

```python
# Read the lines into a list
from io import TextIOWrapper

def read_lines(filename: str) -> list[str]:
    """Read every file name into a list."""
    lines: list[str] = []
    file_handle: TextIOWrapper = open(filename, "r")
    for line in file_handle:
        line = line.lower() # changes the letters to lowercase
        line = line.strip() # removes all of the \n's 
        lines.append(line)
    file_handle.close()
    return lines

shakespeare_lines: list[str] = read_lines("./shakespeare.txt")
print(len(shakespeare_lines))
```

```python
# Tally up characters

# Goal: Define a function named tally
# It will take 2 parameters
# 1. A reference to a dictionary whose keys are str and values are int
# 2. A str that refers to a key in the ditionary you are tallying
# Logic:
# If the str / key is in the dictionary, increase its value by one. 
# Else set the key's value to 1 in the dictionary
# This function returns nothing

from itertools import count


def tally(dictionary1: dict[str, int], key: str) -> None:
    if key in dictionary1:
        dictionary1[key] += 1
    else:
        dictionary1[key] = 1


def count_letters(lines: list[str]) -> dict[str, int]:
    """Counts the frequencies of all letters in lines."""
    counts: dict[str, int] = {}
    for line in shakespeare_lines:
        for char in line:
            if char >= "a" and char <= "z":
                tally(counts, char)
    return counts

shakespeare_letters: dict[str, int] = count_letters(shakespeare_lines)
shakespeare_letters
```

```python
# Sorting the letters

items_in_dict: list[tuple[str, int]] = list(shakespeare_letters.items())
items_in_dict = sorted(items_in_dict)
shakespeare_letters_sorted: dict[str, int] = dict(items_in_dict)
items_in_dict 
```

```python
# Make a bar chart

from matplotlib import pyplot

pyplot.title("Frequencies of Letters in Shakespeare's Works")
pyplot.xlabel("Letters")
pyplot.ylabel("Frequencies")
labels: list[str] = list(shakespeare_letters_sorted.keys()) # list of keys
values: list[int] = list(shakespeare_letters_sorted.values()) # list of values
pyplot.bar(labels, values)
```
